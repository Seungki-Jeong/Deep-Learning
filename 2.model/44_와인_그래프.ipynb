### 그래프로 확인

```python
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
df_pre = pd.read_csv('../data/wine.csv', header=None)
df = df_pre.sample(frac=1)
dataset = df.values
X = dataset[:,0:12]
Y = dataset[:,12]
```

```python
df.head()
```

```python
model = Sequential([
    Dense(30, input_dim=12, activation='relu'),
    Dense(12, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
]) 
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history = model.fit(X, Y, validation_split=0.33, epochs=2000, batch_size=500)
```

```python
y_vloss = history.history['val_loss']
```

```python
y_acc = history.history['accuracy']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize = (12, 8))
plt.plot(x_len, y_vloss, 'o', c = 'red', markersize = 2)
plt.plot(x_len, y_acc, 'o', c = 'blue', markersize = 2)
plt.show()
```

```python
y_vacc = history.history['val_accuracy']
```

```python
y_loss = history.history['loss']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize = (12, 8))
plt.plot(x_len, y_loss, 'o', c = 'red', markersize = 2)
plt.plot(x_len, y_vacc, 'o', c = 'blue', markersize = 2)
plt.plot(x_len, y_vloss, 'o', c = 'orange', markersize = 2)
plt.plot(x_len, y_acc, 'o', c = 'green', markersize = 2)
plt.show()
```
