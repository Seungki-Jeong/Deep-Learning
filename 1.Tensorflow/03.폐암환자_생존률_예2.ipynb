# 폐암 수술 환자의 생존률 데이터 분석

```python
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
seed = 2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
data_set = np.loadtxt("../data/ThoraricSurgery.csv", delimiter = ",")
data_set[:3, :]
```

```python
x_data = data_set[:, 0:17]
y_data = data_set[:, 17]
x_data.shape, y_data.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(24, input_shape = (17,), activation='relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
# 딥러닝 환경 설정(오차함수, 최적화함수)
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam', metrics = ['accuracy'])
```

```python
# 딥러닝 학습
model.fit(x_data, y_data, epochs = 500, batch_size = 30)
```

```python
print("\n Accuracy: %.4f"%(model.evaluate(x_data, y_data, verbose=2))[1])
```
